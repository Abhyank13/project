```python
# Create a Library Management System

# 1. Username Password
# 2. total availables books (math - 0)
# 3. add book
# 4. borrow books
# 5. return books
# 6. display borrowed books details along with person's name. 

# increase and decrease the count of books accordingly.

user="abhi"
password=12345


    


name=input("enter your user name: ")


books={"maths":0,"english":5,"science":10}
borrow={}


if name.lower()==user:
    pd=int(input("enter your password: "))
    if pd==password:
        print("Login Successful")
        
        while True:
                print("-------------------------------------------------------------------------------------------")

                a=int(input("""press 1 to add books.\npress 2 to borrow book\npress 3 to return book\npress 4 to display borrowed book details\npress 5 to log out: """))
                print("-----------------------------------------------------------------------")

                if a==1:
                    book_name=input("Enter the of the book You want to add: ")
                    book_quantity=int(input("How many book do you want add: "))
                    books[book_name]= book_quantity
                    print("-----------------------------------------------------------------------")
                    print(books)
                    continue
                   
                    
                elif a==2:
                    name=input("enter your name: ")
                    borrow_name=input("Enter book name you want to borrow: ")
                    print("-----------------------------------------------------------------------")
                    borrow[name]=borrow_name
                    books[borrow_name]-=1
                    print("-----------------------------------------------------------------------")
                    print(books)
                    print(borrow)
                    print("-----------------------------------------------------------------------")
                   
                    continue

                elif a==3:
                    borrowed_book=input("enter the book name you borrowed: ")
                    r=input("enter the name: ")
                    books[borrowed_book]+=1
                    del borrow[r]
                    print("-----------------------------------------------------------------------")
                    print(books)
                    print(borrow)
                    print("-----------------------------------------------------------------------")

                    continue
                elif a==4:
                    print("-----------------------------------------------------------------------")
                    print(borrow)
                    print("-----------------------------------------------------------------------")
                    
                    continue
                    
                elif a==5:
                    print("Thank you for visiting.")
                    break
                    
             
                else:
                    print("invalid password!")
else:
    print("invalid username")
    

```
